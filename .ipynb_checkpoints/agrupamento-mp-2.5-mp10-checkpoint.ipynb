```python
from sklearn.preprocessing import StandardScaler
std_data = StandardScaler().fit_transform(dados)
kmeans = KMeans(n_clusters=3, init='k-means++', n_init=10, random_state=1234)
kmeans.fit(std_data)
C = kmeans.cluster_centers_

plt.rcParams['figure.figsize'] = (16, 9)

# Creating a sample dataset with 4 clusters
#X, y = make_blobs(n_samples=800, n_features=3, centers=4)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(std_data[:, 0], std_data[:, 1], std_data[:, 2],  c = kmeans.labels_, cmap = "rainbow")
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c='#050505', s=1000)

C
```

```python
# !! Get the indices of the points for each corresponding cluster
mydict = {i: np.where(kmeans.labels_ == i)[0] for i in range(kmeans.n_clusters)}
```

```python
mydict
```

```python
dataset = pd.read_csv('Cidades-mp-2.5-novo.csv',sep=';',encoding = "ISO-8859-1")
for j in mydict:

    nome_arquivo='Grupo-mp-2.5'+str(j)+'.txt'
    print (nome_arquivo)
    arquivo = open(nome_arquivo, "w")
   
    for m in mydict[j]:
        print(dataset.iloc[m][0])
        arquivo.write('\n')
        arquivo.write(dataset.iloc[m][0])
        
    arquivo.close()
```

# Comparativo com o K-medoids

```python

from pyclustering.cluster.kmedoids import kmedoids


from pyclustering.cluster import cluster_visualizer
std_data = StandardScaler().fit_transform(dados)
std_data
cluster = kmedoids(std_data[:, 0:24], [3, 12, 20])


cluster.get_medoids()

cluster.process()

previsoes = cluster.get_clusters()


medoides = cluster.get_medoids()



v = cluster_visualizer()

v.append_clusters(previsoes, std_data[:,0:2])

v.append_cluster(medoides, data = std_data[:,0:2], marker = '*', markersize = 15)


v.show()

previsoes


#medoides


```

```python
##Criar grafico de barras
arrayClusters=[0] * 21

for i in range(len(previsoes)):
    for j in previsoes[i]:
        arrayClusters[j]=i;

dados_Limpeza["clustersK-medoids"] = arrayClusters
dados_Limpeza.groupby("clustersK-medoids").aggregate("mean").plot(kind = 'bar', figsize=(20, 10), fontsize = 10)
dados_Limpeza["clustersK-medoids"]
```

## Escrevendo os grupos 

```python
dataset = pd.read_csv('Cidades-mp-2.5-novo.csv',sep=';',encoding = "ISO-8859-1")
for j in range(len(previsoes)):

    nome_arquivo='K-medoids-Grupo-'+str(j)+'-mp-2.5.txt'
    print (nome_arquivo)
    arquivo = open(nome_arquivo, "w")
   
    for m in previsoes[j]:
        
        arquivo.write('\n')
        arquivo.write(dataset.iloc[m][0])
        
    arquivo.close()
```

# Agrupamento MP10

```python
dadosMp10 = pd.read_csv('dataset-mp10.csv',sep=';',encoding = "ISO-8859-1")
.head()
```

```python
#Apresenta a quantidade de valores nulos por mes
dadosMp10.isna().sum()
```

```python
##Remoção de valores nullos ou vazios
periodomp10=['jan_2017', 'fev_2017', 'mar_2017', 'abr_2017', 'mai_2017',
       'jun_2017', 'jul_2017', 'ago_2017', 'set_2017', 'out_2017', 'nov_2017',
       'dez_2017', 'jan_2018', 'fev_2018', 'mar_2018', 'abr_2018', 'mai_2018',
       'jun_2018', 'jul_2018', 'ago_2018', 'set_2018', 'out_2018', 'nov_2018',
       'dez_2018']
for i in range(len(periodomp10)):
    #tenho que calcular a media da coluna que não seja cidade
    valor_media_mp10=math.floor(dadosMp10[periodomp10[i]].mean()) 
    dadosMp10.update(dadosMp10[periodomp10[i]].fillna(valor_media_mp10))
```

```python
##Removendo o atributo cidade para realizar normalização dos dados
dados_LimpezaMp10=dadosMp10

dados_LimpezaMp10.head()
##Normalizando os valores

dadosNormalizados_mp10 = StandardScaler().fit_transform(dados_LimpezaMp10)
```

## Aplicando K-means sobre os dados Mp 10

```python
#encontrando o melhor número de k
inertiaMp10 = []

for i in range(1,11):
    kmeans_mp10 = KMeans(n_clusters=i, random_state=1234)
    kmeans_mp10.fit(dadosNormalizados_mp10)
    inertiaMp10.append((i,kmeans_mp10.inertia_,))
plt.plot([w[0] for w in inertiaMp10],[w[1] for w in inertiaMp10], marker="X")
```

```python
##O melhor valor de k =3
kmeans_mp10 = KMeans(n_clusters=3, init='k-means++', n_init=10, random_state=1234)
kmeans_mp10.fit(dadosNormalizados_mp10)
cluster_list_mp10 = kmeans_mp10.labels_

##gerar vizualização

dados_LimpezaMp10["clusters"] = cluster_list_mp10
dados_LimpezaMp10.groupby("clusters").aggregate("mean").plot(kind = 'bar', figsize=(20, 10), fontsize = 10)
```

```python
C_mp10 = kmeans_mp10.cluster_centers_

plt.rcParams['figure.figsize'] = (16, 9)

# Creating a sample dataset with 4 clusters
#X, y = make_blobs(n_samples=800, n_features=3, centers=4)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(dadosNormalizados_mp10[:, 0], dadosNormalizados_mp10[:, 1], dadosNormalizados_mp10[:, 2],  c = kmeans_mp10.labels_, cmap = "rainbow")
ax.scatter(C_mp10[:, 0], C_mp10[:, 1], C_mp10[:, 2], marker='*', c='#050505', s=1000)
```

```python
##Escrevendo em um arquivo os grupos gerados

mydict_mp10 = {i: np.where(kmeans_mp10.labels_ == i)[0] for i in range(kmeans_mp10.n_clusters)}
datasetMp10=pd.read_csv('dataset-mp10.csv',sep=';',encoding = "ISO-8859-1")
for j in mydict_mp10:

    nome_arquivo='Grupo-mp10-'+str(j)+'.txt'
   
    arquivo = open(nome_arquivo, "w")
   
    for m in mydict_mp10[j]:
        
        arquivo.write('\n')
        arquivo.write(datasetMp10.iloc[m][0])
        
    arquivo.close()
```
